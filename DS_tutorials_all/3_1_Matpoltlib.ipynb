# Matplotlib

## Introduction

It was created by John Hunter. He created it to try to replicate MatLab's (another programming language) plotting capabilities in Python.

It is an excellent 2D and 3D graphics library for generating scientific figures. 

Some of the major Pros of Matplotlib are:

* Generally easy to get started for simple plots
* Support for custom labels and texts
* Great control of every element in a figure
* High-quality output in many formats
* Very customizable in general

Matplotlib allows you to create different type of figures programmatically.

To have a close look at matplotlib

visit matplotlib official web page: http://matplotlib.org/

### Plotting two lines and Customizations of Line Graph

```python
x,y = sorted(np.random.rand(5)),sorted(np.random.rand(5))

plt.plot(x , y , 
         color = 'green' ,
         label='random values' ,
         linestyle = ':' ,
         linewidth = 2 ,
         marker = '*' ,
         markersize=25
        )

plt.xlabel('X Axis Name' , color = 'blue' , size = 20)   # x axis label
plt.ylabel('Y Axis Name')                             # y axis label
plt.title('Graph title',size = 30) 

plt.legend();
```

```python
x,y = sorted(np.random.rand(5)),sorted(np.random.rand(5))

x1,y1 = sorted(np.random.rand(5)),sorted(np.random.rand(5))

# line 1
plt.plot(x , y , 
         color = 'green' ,
         label='line 1' ,
         linestyle = '--' ,
         linewidth = 2 ,
         marker = 'o' ,
         markersize=15)
# label is mentioned here for plt.legend()


# line 2
plt.plot( x1 , y1 , 'b*-', label='line 2' ,
         markersize=15 ,
         linewidth = 3 ,
         markerfacecolor = 'red',
         markeredgecolor = 'yellow' ,
         markeredgewidth=1)
#  b*-  = is the shortcode for , marker , color , linestyle
# marker (*) with b (blue as colour) ,line style (-) ,'b*-' is called fmt parameter
# markerfacecolor = color of marker , markeredgecolor = markers edges color
# label is mentioned here for plt.legend()


plt.xlabel('X Axis Name' , color = 'blue' , size = 20)           # x axis label
plt.ylabel('Y Axis Name')                                        # y axis label
plt.title('Graph title',size = 30) 

plt.legend()  # if de don't mention this line then labels of lines will not show up.

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
# possible linestype options ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
# we can use the linewidth or lw keyword argument. The line style can be selected using the linestyle or ls keyword arguments
```

### Scatter plot

plt**.scatter**() gives scatter plot

**Syntax**
 plt.scatter(x ,y, c=None ,marker=None, alpha=None, linewidths=None, edgecolors=None , s)
 
* c = color


* s = size


* alpha = transparency in colours   (0 (transparent) and 1 (opaque))


* edgecolors = color or sequence of color, optional ,The edge color of the marker

```python
x,y,x1,y1 = np.random.randint(1,20,10),np.random.randint(1,20,10),np.random.randint(1,20,10),np.random.randint(1,20,10)

plt.scatter(x , y ,label='line 1' ) 

plt.scatter(x1 , y1,  label='line 2', marker = '^' ,
            edgecolors= 'black' , color = 'orange', s = 400 ,alpha = 0.5 ) 

plt.xlabel('X Axis Name' , color = 'blue' , size = 20)           # x axis label
plt.ylabel('Y Axis Name')
plt.title('Graph title',color = 'green',size = 30) 
plt.legend()  
```

### Pie Chart

```python
parts = [7,2,2,3]
activity = ['sale','manufacture','manpower','maintaince']
colours_we_want = ['#ff9999','#66b3ff','#99ff99','#ffcc99']    # these are  RGB hex colour codes for colours
                                                               # we cannot use alpha so use lighr colours by hex code
                                                      
plt.pie( parts ,
        labels = activity ,
        explode = (0,0,0.2,0),       # explode slices out a particular slice , brings it out , here 3rd slice is tsken out.     
        colors = colours_we_want ,      
        autopct = '%1.1f%%',         # autopct shows percent weight of slice ,  we have to learn the string code
        startangle = 90 ,            # startangle makes the starting aligned at 90 degree and then moves counter clockwise
        shadow = True ,              # shadow displays the shadow of slices
        radius = 2 ,                 # for radius of the circle
        counterclock=True ,          # labeling done counter clockwise
        center=(0, 0), frame=False , # frame is used for displaying graph if done true.     
        textprops={'fontsize': 24} )                         
```

### Bar Graph

plt**.bar**() gives bar graph

**Syntax**
plt.bar(x, height = y, width=0.8, align='center', color , edgecolor  ,alpha)

align :{'center', 'edge'}, optional, default: 'center', Alignment of the bars to the *x* coordinates

```python
x,y,x1,y1 = [1,2,3],[2,5,7],[5,7,10],[1,2,3]

plt.bar(x,y,  label='bars 1')

plt.bar(x1,y1,  label='bars 2' ,  width=0.8 ,alpha = 0.3  )

plt.legend()
```

```python
x,y,x1,y1 = [10,20,30],[2,5,7],[5,7,10],[1,2,3]

plt.bar(x,y,  label='bars 1', width=1.9)

plt.bar(x1,y1,  label='bars 2' ,  
        width=1.8 ,
        alpha = 0.7,
        color = 'orange',
        align = 'edge')

plt.legend()
```

# Matplotlib Object Oriented Method
Matplotlib's Object Oriented API. This helps to instantiate figure objects and then call methods or attributes from that object.

This approach is nicer when dealing with a canvas that has multiple plots on it. 

To begin we create a figure instance. Then we can add axes to that figure:

```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
axes.plot(x, y, 'b')  # b = blue
axes.set_xlabel('Set X Label') # Notice the use of set_ to begin methods
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
```

```python
# Creates blank canvas
fig = plt.figure(figsize =(3,3) )

axes1 = fig.add_axes([0, 0, 1, 1]) # main axes            # left, bottom, width, height (range 0 to 1)
axes2 = fig.add_axes([1.3, 0 , 0.5, 0.5]) # inset axes    # left, bottom, width, height (range 0 to 1)

# Larger Figure Axes 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes1')
axes1.set_ylabel('Y_label_axes1')
axes1.set_title('Axes 1 Title')

# Insert Figure Axes 2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title')

fig.tight_layout();
```

A common issue with matplolib is overlapping subplots or figures. We ca use **fig.tight_layout()** or **plt.tight_layout()** method, which automatically adjusts the positions of the axes on the figure canvas so that there is no overlapping content:

## SUBPOTS

```python
# plt.subplot(number of rows, number of rows cols, plot_number)
plt.subplot(1,2,1)
plt.plot(x, y)

plt.subplot(1,2,2)
plt.plot(y, x);

plt.tight_layout()   # fixes the overlapping of graphs
```

```python
# plt.subplot(number of rows, number of rows cols, plot_number)
# numbering of subplots starts from top left to right and than next row 

plt.subplot(3,3,1)    #total subplots = 3*3 = 9
plt.plot(x, y)

plt.subplot(3,3,5)
plt.scatter(y, x);

plt.tight_layout()   # fixes the overlapping of graphs
```

### Figure size, aspect ratio and DPI

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the `figsize` and `dpi` keyword arguments. 
* `figsize` is a tuple of the width and height of the figure in inches
* `dpi` is the dots-per-inch (pixel per inch). 

For example:

```python
fig = plt.figure(figsize=(8,4), dpi=100)
```

The same arguments can also be passed to layout managers, such as the `subplots` function:

```python
fig, axes = plt.subplots(figsize=(12,3))

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

## Saving figures
Matplotlib can generate high-quality output in a number formats, including PNG, JPG, EPS, SVG, PGF and PDF. 

To save a figure to a file we can use the `savefig` method in the `Figure` class:

```python
fig.savefig("filename.png")
```

Here we can also optionally specify the DPI and choose between different output formats:

```python
fig.savefig("filename.png", dpi=200)
```
